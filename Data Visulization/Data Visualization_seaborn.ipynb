```python
import numpy as np
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

%matplotlib inline
```

# Bar plots

```python
tips = sns.load_dataset('tips')
```

```python
sns.barplot(x='day', y='tip',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
#to change color
sns.barplot(x='day',y='total_bill',data=tips,hue='sex',palette='winter_r')
```

```python
#smoker
sns.barplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.barplot(x='total_bill',y='day',data=tips,hue='sex',palette='spring')
```

# Distplot

```python
#we are using numpy to create the random numbers
num= np.random.randn(150)
```

```python
sns.distplot(num)
```

```python
sns.distplot(num,color='red')
```

```python
label_dist=pd.Series(num,name="variable x")
```

```python
sns.distplot(label_dist)
```
